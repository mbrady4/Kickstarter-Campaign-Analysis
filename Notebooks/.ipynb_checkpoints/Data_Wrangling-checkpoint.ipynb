## Process to import the files and form a master dataframe for each month

- This step is necessary due to the file storage format which divides each monthly scrape into 56 separate CSV files. Fortunately there is a standard file naming convention. 
- A process similiar to this will likely be repeated for each month

```python
df = pd.read_csv('Kickstarter_2019-03-14T03_20_12_200Z/Kickstarter.csv')
df.head()
```

```python
df.shape
```

```python
appended_df = df.copy()
```

```python
for i in range(1, 56): 
    file = ''
    if i < 10:
        file = 'Kickstarter_2019-03-14T03_20_12_200Z/Kickstarter00' + str(i) + '.csv'
    else: 
        file = 'Kickstarter_2019-03-14T03_20_12_200Z/Kickstarter0' + str(i) + '.csv'
    temp = pd.read_csv(file)
    appended_df = appended_df.append(temp)
    
appended_df.head()
```

```python
appended_df.shape
```

```python
appended_df['slug'].nunique()
```

### Investigating usd_type missing values

- Check the count of values to determine if it would be appropriate to fill with one value

```python
# Nearly an even split between international and domestic! 
master['usd_type'].value_counts()
```

```python
# Given the even split, let's test how a forward fill would impact the ratio
master['usd_type'].ffill().value_counts()
```

### Converting columns with strings containing dictionaries to tidy columns
- As you likely have noticed, several of our columns contain information that appears to be in the form of a dictionary but is currently not easily analyzed. 
- In this step we will expand all of the dictionary columns, so that each key has its own column (note: creating a function to handle this step would be better than my method below)

### Category Column

```python
temp = pd.DataFrame(master['category'])
temp.head()
```

```python
# Notice that the values within the column are stored as strings but could be interpretted as dictionaries
temp['category'][0]
```

```python
# Get pandas to evaluate the values as a dictionary instead of a string
import ast
temp['category'] = temp['category'].apply(ast.literal_eval)
```

```python
# Convert each dictionary key to a column with the coresponding values
temp = temp['category'].apply(pd.Series)
```

```python
# Rename columns to avoid confusion and duplication when combined with back with master DataFrame
temp = temp.rename(columns={
        'id':'cat_id',
        'name':'cat_name',
        'slug':'cat_slug',
        'position':'cat_position',
        'parent_id':'cat_parent_id',
        'color':'cat_color',
        'urls':'cat_urls'
        })
```

```python
temp.head()
```

```python
master = pd.concat([master.drop(['category'], axis=1), temp], axis=1)
```

```python
master.head()
```

```python
master['cat_urls'][0]
```

### Location Column
- Follow similiar process to the category column

```python
x = pd.DataFrame(master['location'])

# It appears that we likely don't want to add all of these attributes to our master DataFrame
x.head()
```

```python
x['location'][0]
```

```python
x.dtypes
```

```python
def turn_to_dict(string):
    d = dict()
    attributes = str(string).split(',')
    for attribute in attributes:
        if ':' in attribute:
            key, value = attribute.split(':', maxsplit=1)
            key = key.strip('"{')
            value = value.strip('"')
            d[key] = value
    return d
```

```python
x['location'] = x['location'].apply(turn_to_dict)
```

```python
x = x['location'].apply(pd.Series)
```

```python
x.head()
```

```python
x = x.drop(columns=['urls', 'location', 'api'])
x = x.rename(columns={
    'id':'loc_id',
    'name':'loc_name',
    'slug':'loc_slug',
    'short_name':'loc_short_name',
    'displayable_name':'loc_display_name',
    'localized_name':'loc_local_name',
    'country':'loc_country',
    'state':'loc_state',
    'type':'loc_type',
    'is_root':'loc_is_root'
})
```

```python
x.head()
```

```python
master = pd.concat([master.drop(['location'], axis=1), x], axis=1)
```

```python
master.head()
```
