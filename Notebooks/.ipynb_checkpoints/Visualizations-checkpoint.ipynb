### Dependencies

### Importing Dataset

```python
df = pd.read_csv('../Compiled_Campaign_Data.csv')
```

```python
df = df.drop_duplicates(['project_id'], keep='first')
```

```python
df.shape
```

```python
df['project_id'].nunique()
```

### Engineering Features

Creating a row to record the launch date of campaigns

```python
df['created_at'] = pd.to_datetime(df['created_at'])
```

```python
labels = [2009, 2010, 2011, 2012, 2013, 2014, 
          2015, 2016, 2017, 2018, 2019]
date_bins = pd.date_range('2008-01-01 23:59:59', 
                          freq='1Y', 
                          periods=12)
df['created_year'] = pd.cut(df['created_at'], 
                            date_bins, 
                            labels=labels)
```

Creating a column to record the usd_per_pledge

```python
df['usd_per_pledge'] = ( df['usd_pledged'] / df['backers_count'] )
```

Creating a binary column which denotes whether a campaign is succssful or not (1=successful, 0=failed/suspended/cancelled, np.nan=live)

```python
df['bin_state'] = df['state']
df['bin_state'] = df['bin_state'].replace({
    'successful': 1,
    'live': np.nan,
    'failed': 0,
    'suspended': 0,
    'canceled': 0
})
```

### Subsetting the Dataset

Creating a dataset with just successful campaigns

```python
successful = df[ df['state'] == 'successful' ]
```

```python
successful.shape
```

Creating a dataframe with just US-based campaigns

```python
us = df[ df['loc_country'] == 'US' ]
us['loc_state'].nunique()
```

### Visualizing the change in campaigns over time

```python
year_counts = df[ df['created_year'] > 2013]['created_year'].value_counts()[:6]
year_counts = year_counts.sort_index()[:5]
```

Now that we have the data we want to visualize, we will use matplotlib

```python
# Set plot style
plt.style.use('fivethirtyeight')

# Adjust figure size
fig, ax = plt.subplots(figsize=(7.25,7.25))

# Adjust color of gridlines
ax.grid(alpha=0.7, linewidth=1.25)

# Add title
ax.text(x=-4000,
        y=4.85, 
        s="The Number of Campaigns on Kickstarter has Stagnated",
        fontsize=18,
        fontweight='bold')
ax.text(x=-4000,
        y=4.58, 
        s="Post 2015, Annual campaigns have stabalized ~27-28K",
        fontsize=15,
        fontweight='normal')

# Data

# Create bar chart
year_counts.plot.barh(ax=ax, 
                     color='#1d8ed2')

# Add labels to bars
for i, v in enumerate(year_counts):
    ax.text(v - 5000, i-0.04, "{:,}".format(int(v)), color='white', fontweight='bold')

# Set location of x-axis ticks
ax.set(xticks=[0, 10000, 20000, 30000, 40000]
      )

# Set x-axis tick labels (not strictly neccesary in this case)
ax.set_xticklabels( ["0", "10,000", "20,000", "30,000", "40,000"] )

# Set x-axis label
ax.set_xlabel('Total Number of Campaigns', fontsize=13)

# Set font-size of axes
ax.tick_params(axis='both', labelsize=13)

plt.savefig('number_of_campaigns.png',)

# Show plot without extra output
plt.show()
```

### Visualizing the success rate of campaigns based on origin

```python
import altair as alt
from vega_datasets import data
alt.renderers.enable('notebook')

states = alt.topo_feature(data.us_10m.url, 'states')
source = data.population_engineers_hurricanes.url
variable_list = ['population', 'engineers', 'hurricanes']

alt.Chart(states).mark_geoshape().encode(
    alt.Color(alt.repeat('row'), type='quantitative')
).transform_lookup(
    lookup='id',
    from_=alt.LookupData(source, 'id', variable_list)
).properties(
    width=500,
    height=300
).project(
    type='albersUsa'
).repeat(
    row=variable_list
).resolve_scale(
    color='independent'
)
```

```python
state_counts = us['loc_state'].value_counts(ascending=False, normalize=True)
```

```python
state_campaign_counts = ( ( us.groupby('loc_state')['converted_pledged_amount'].sum())
/ ( us.groupby('loc_state')['converted_pledged_amount'].sum()).sum() )
```

```python
state_success.columns.tolist()
```

```python
state_counts = state_counts.to_frame()
```

```python
state_counts = state_counts.reset_index()
state_counts = state_counts.rename(columns={
  'loc_state':'Percent of US-Based Campaigns',
  'index':'loc_state'
})
```

```python
state_counts.columns.tolist()
```

```python
states = pd.merge(state_success, state_counts, on='loc_state')
```

```python
state_campaign_counts = state_campaign_counts.to_frame()
```

```python
state_campaign_counts = state_campaign_counts.reset_index()
```

```python
state_campaign_counts.columns.tolist()
```

```python
state_campaign_counts.head()
```

```python
state_campaign_counts = state_campaign_counts.rename(columns={
    'converted_pledged_amount':'Share of total contributions to US Campaigns'
})
```

```python
states = pd.merge(states, state_campaign_counts, on='loc_state')
```

```python
states.head()
```

```python
states = states.drop(columns=[0.0, 'level_0'])
```

```python
states = states.rename(columns={
    'loc_state':'State',
    1.0:'Success Rate of Campaigns'
})
```

```python
states['Success Rate of Campaigns'] = pd.Series(["{0:.2f}%".format(val * 100) for val in states['Success Rate of Campaigns']], index = states.index)
```

```python
states['Percent of US-Based Campaigns'] = pd.Series(["{0:.2f}%".format(val * 100) for val in states['Percent of US-Based Campaigns']], index = states.index)
```

```python
states['Share of total contributions to US Campaigns'] = pd.Series(["{0:.2f}%".format(val * 100) for val in states['Share of total contributions to US Campaigns']], index = states.index)
```

```python
states.head()
```

```python
states.shape
```

```python
states = states.sort_values('Success Rate of Campaigns', ascending=False)[:10]
```
